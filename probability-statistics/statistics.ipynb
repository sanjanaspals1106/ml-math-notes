```python
import numpy as np
import matplotlib.pyplot as plt
data=np.random.normal(loc=50, scale=10, size=1000)
```

```python
#Visualizing the data
plt.hist(data, bins=30, color="pink")
plt.title("Data distribution")
plt.show()
```

```python
print(np.mean(data))
print(np.std(data))
print(np.var(data))
```

```python
data2=np.random.normal(loc=50, scale=3, size=1000)
plt.hist(data, bins=30, color="pink", label="More Noisy")
plt.hist(data2, bins=30, color="skyblue", label="Less Noisy")
plt.legend()
plt.show()
```

Real-world data contains noise and variability.
Statistics helps us understand the overall behavior of data rather than individual samples.
Because data is noisy, a model cannot achieve zero loss.
Instead, training focuses on minimizing average error, not perfect prediction.
