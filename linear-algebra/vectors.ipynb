```python
a=[1,2,3]
b=[4,5,6]
```

```python
#dot product using for loop
dot=0
for i in range(len(a)):
    dot+=a[i]*b[i]
print(dot)
```

```python
#dot product using numpy
import numpy as np
np.dot(a,b)
```

```python
x=np.array([2,3])
w1=np.array([1,1])
w2=np.array([5,-1])

print(np.dot(x,w1))
print(np.dot(x,w2))
```

Each element of a vector represents a feature value.
The weight vector represents the importance of each feature.
The dot product combines the values into a single score.
Larger weight means stronger influence.
The output value is used to make predictions.
Linear regression is the sum of dot product plus bias.
