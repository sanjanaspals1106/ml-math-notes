```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_true=np.array([3,5,7,9])
y_good_pred=np.array([2.8,5.1,7.2,8.9])
y_bad_pred=np.array([0,0,0,0])
```

```python
def MAE(true, pred):
    return np.mean(abs(true-pred))
print(MAE(y_true, y_good_pred))
print(MAE(y_true, y_bad_pred))
```

```python
def MSE(true, pred):
    return np.mean((true-pred)**2)
print(MSE(y_true, y_good_pred))
print(MSE(y_true, y_bad_pred))
```

```python
#Visualising loss functions
errors=np.linspace(-10,10,100)
loss=errors**2
plt.plot(errors, loss, color="pink")
plt.xlabel("Error")
plt.ylabel("Loss")
plt.title("MSE curve")
plt.show()
```

Loss Function:
A loss function measures how wrong a modelâ€™s prediction is for a single data point.
It converts prediction error into a numeric value that the model can optimize.
Lower loss indicates better predictions.
Loss vs Cost
Loss: error for one training example
Cost: average loss over the entire dataset
During training, models aim to minimize the cost, not individual errors.

Mean Squared Error (MSE):
Mean Squared Error is a commonly used loss function for regression problems.
It works by:
computing the difference between predicted and true values
squaring the error to penalize large mistakes
averaging over all samples
Squaring the error makes MSE sensitive to outliers, as large errors contribute disproportionately to the loss.
Effect of Outliers
Outliers are data points that lie far from the general pattern of the data.
Because MSE squares the error, outliers can significantly influence the loss and affect model training.


In classification problems, models often output probabilities instead of direct numeric values.
Loss functions such as log loss evaluate how well these probabilities match the true class labels.
A key idea of log loss is:
confident correct predictions are rewarded
confident wrong predictions are penalized heavily
This encourages models to not only predict the correct class, but also to be appropriately confident.
