```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2
```

```python
x=np.linspace(-5, 5, 100)
y=f(x)

plt.plot(x,y)
plt.axhline(0)
plt.axvline(0)
plt.title("y=x^2")
plt.show()
```

```python
def derivative(f, x, h=1e-4):
    return (f(x+h)-f(x))/h
```

```python
for p in [-3, -1, 0, 1, 3]:
    print(p, derivative(f, p))
```

```python
x0=2
slope=derivative(f,x0)

y0=f(x0)
tangent=slope*(x-x0)+y0
plt.plot(x,y,label="y=x^2")
plt.plot(x,tangent,'--', label="tangent at x=2")
plt.scatter([x0],[y0])
plt.legend()
plt.show()
```

A derivative measures how fast a function changes.
Slope tells the direction of change.
Positive slope means increasing, negative means decreasing.
Zero slope means no change.
Optimization uses slope to decide which direction to move.
Learning means adjusting values using slope information.
