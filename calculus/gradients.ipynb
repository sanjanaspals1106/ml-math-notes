```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(w1,w2):
    return w1**2+w2**2
```

```python
w1, w2 = 0.5, 2.5
print(loss(w1,w2))
```

```python
epsilon=1e-4

dw1=(loss(w1+epsilon, w2)-loss(w1,w2))/epsilon
dw2=(loss(w1, w2+epsilon)-loss(w1,w2))/epsilon

gradient=np.array([dw1,dw2])
print(gradient)
```

```python
learning_rate=0.1
w1_new=w1-learning_rate*w1
w2_new=w2-learning_rate*w2

print(f"Old loss: {loss(w1,w2)}")
print(f"New loss: {loss(w1_new,w2_new)}")
```

```python
w1_vals = np.linspace(-3, 3, 100)
w2_vals = np.linspace(-3, 3, 100)

W1, W2 = np.meshgrid(w1_vals, w2_vals)
Z = loss(W1, W2)

plt.contour(W1, W2, Z, levels=20)
plt.quiver(w1, w2, -dw1, -dw2, color='red')
plt.scatter(0, 0)
plt.title("Gradient Descent Direction")
plt.show()
```

Gradient is the vector of partial derivatives of the loss with respect to each weight.
It points in the direction of the steepest increase of the loss.
To minimize the loss, we move in the opposite direction of the gradient (gradient descent).
Since each component of the gradient represents the slope of the loss with respect to a particular weight,
the gradient collectively tells the model how all weights should be updated to reduce the loss.
